```python
import pandas as pd
import plotly.express as px

data = pd.read_csv('covid-data.csv')
fig = px.line(data, x="total_cases", y="new_cases",title='Nepal Covid New Cases over total cases | Data by Shaswot Lamichhane')
fig['data'][0]['line']['color'] = '#f70254 '
fig.show()
```

```python
df = px.data.gapminder().query("year == 2007").query("continent == 'Asia'")
df
df.loc[df['pop'] < 24.e6, 'country'] = 'Other countries' # Represent only large countries
fig = px.pie(df, values='pop', names='country', title='Nepal Covid Cases | Data by Shaswot Lamichhane',color_discrete_sequence=px.colors.sequential.Aggrnyl)
fig.show()
```
